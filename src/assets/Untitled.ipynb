```python
import pandas as pd
df=pd.read_csv('DummyData.csv')
df=df.drop('Unnamed: 0',axis=1)
df.head()
```

```python
col1= df.columns
cols=[]

for item in col1[:]:
    cols.append('_'.join(item.split()))
cols=['id'] + cols
cols
```

```python
l=696
with open('Data.json',mode='w+') as f:
    f.seek(0)
    f.write("[\n")
    for item in df[:l].itertuples():
        f.write("\t{\n")
        for i in range(len(item)):
            a="{}".format(item[i])
            a=' '.join(a.split("\""))
            if '\n' in a:
                print('This line contain line breaker it needs ti be fixed : ',a)
                a=' '.join(a.split("\n"))
                print('fixed : ', a)
            try:
                if i!= 93:
                    f.write('\t\t\"'+cols[i]+"\": \""+a+"\",\n")
                else:
                    f.write('\t\t\"'+cols[i]+"\": \""+a+"\"\n")
            except:
                print(a,'\n')
                a='720p Webcam'
                f.write('\t\t\"'+cols[i]+"\": \""+a+"\",\n")
                continue
        if item[0] != l-1:
            f.write("\t},\n")
        else :
            f.write("\t}\n")
    f.write("\n]")
```
