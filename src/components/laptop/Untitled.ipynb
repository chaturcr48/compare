```python
import pandas as pd
```

```python
df=pd.read_csv('processed_data.csv')
```

```python
col1= df.columns
cols=[]

for item in col1[:]:
    cols.append('_'.join(item.split()))
cols=['id'] + cols
```

```python
len(df)
```

```python
l=len(df)
with open('finalData.json',mode='w+') as f:
    f.seek(0)
    f.write("[\n")
    for item in df[:l].itertuples():
        f.write("\t{\n")
        for i in range(len(item)):
            a="{}".format(item[i])
            a=' '.join(a.split("\""))
            if '\n' in a:
                print('This line contain line breaker it needs ti be fixed : ',a)
                a=' '.join(a.split("\n"))
                print('fixed : ', a)
            try:
                f.write('\t\t\"'+cols[i]+"\": \""+a)
            except:
                print(cols[i],'\n')
                a=' '
                f.write('\t\t\"'+cols[i]+"\": \""+a)
            if i!= len(cols)-1:
                f.write("\",\n")
            else:
                f.write("\"\n")
        if item[0] != l-1:
            f.write("\t},\n")
        else :
            f.write("\t}\n")
    f.write("\n]")
```
